Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
Answer:
significance level(alpha) is a threshold we set before doing a hypothesis test
It represents the probability of making type-1 error- rejecting the null hypothesis when it actually True

*Common values of alpha are 0.05 (5%), 0.01 (1%), or 0.10 (10%).
It defines the critical region for the test — if the p-value is less than α, we reject the null hypothesis.

Example:
If alpha = 0.05, there is a 5% risk of rejecting the null hypothesis when it is actually true.

* It is a Descision cutoff that balance the risk of making errors in hypothesis

````python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
# 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
# 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
# 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# (Include your Python code and output in the code box below.)
# Answer:

import numpy as np
from scipy import stats


#null hypothesis parameters
population_mean = 50 #Hypothesied mean

#sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
sample_size = len(sample_data) #sample size
sample_mean = np.mean(sample_data) #mean
sample_std = np.std(sample_data, ddof =1) #sample std(unbaised)
std_err = sample_std/np.sqrt(sample_size)

z_score = (sample_mean - population_mean)/std_err

#compute pvalue (2 taild test)
p_value = 2 * (1- stats.norm.cdf(abs(z_score)))

print(f"sample size: {sample_size}")
print(f"sample mean: {sample_mean}")
print(f"standard dev: {sample_std}")
print(f"standard error {std_err}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

#decision
alpha = 0.5

if p_value < alpha:
  print('Reject the null hypothesis')
else:
  print('Fail to reject the null hypothesis')
````

````python
# Question 8: Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# (Include your Python code and output in the code box below.)
# Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#stimulate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10, n=1000

#calculate mean and standard deviation
mean = np.mean(data)
std = np.std(data, ddof=1)
n= len(data)

#compute 95% confidence interval
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1-alpha/2) #z critical
margin_of_error = z_critical * (std/np.sqrt(n))
confidence_interval = (mean-margin_of_error, mean+margin_of_error)

print(f"sample mean: {mean: 2f}")
print(f"sample std: {std}")
print(f"confidence interval: {confidence_interval}")


plt.figure(figsize=(8,5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.5)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label= f"Mean = {mean: 2f}")
plt.axvline(confidence_interval[0], color='green', linestyle='dashed', linewidth=2, label= f"lower bound = {confidence_interval[0]: 2f}")
plt.axvline(confidence_interval[1], color='green', linestyle='dashed', linewidth=2, label= f"upper bound = {confidence_interval[1]: 2f}")

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

````python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.
# (Include your Python code and output in the code box below.)
# Answer:

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
  #calculate z score for data set
  # z= (x- mean)/ std
  mean = np.mean(data)
  std = np.std(data, ddof=1)
  z_scores = [(x-mean)/std for x in data]
  return z_scores, mean, std


#example data set

np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10, n=1000

#calculate z-score

z_score, mean,std = calculate_z_scores(data)

print(f"Original mean ", mean)
print(f"standard devation ", std)

print("Histogram Standard deviation z scores")
#plot histogram
plt.figure(figsize=(8,5))
plt.hist(z_score, bins=15, edgecolor='black', alpha=0.5)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label= f"Mean = 0")
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````
