```
Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.
```

```
Question 2: What are the key assumptions of Simple Linear Regression?
Answer:
Simple Linear Regression (SLR) works correctly only when the following assumptions are satisfied:

1. Linearity: There must be a linear relationship between the independent variable (X) and dependent variable (Y).
The data should follow a straight-line trend.

2. Independence of Errors: The residuals (errors) must be independent of each other.
No correlation should exist between error terms.

3. Homoscedasticity (Constant Variance): The variance of residuals should be constant across all levels of X.
Points should be evenly spread around the regression line.

4. Normality of Errors: The residuals should follow a normal distribution.
This is important for hypothesis testing and confidence intervals.

5. No Outliers / High Leverage Points: Extreme values can distort the
regression line, so data should not have influential outliers.
```

```
Question 5: What is the method of least squares in linear regression?
Answer:The method of least squares is a technique used in linear regression to find the best-fitting line through a set of data points.

It works by minimizing the sum of the squared differences between the actual values and the predicted values.

These differences are called residuals:
Residual=(Yacutual - Ypredicted)
The least squares method chooses the values of Œ≤‚ÇÄ (intercept) and Œ≤‚ÇÅ (slope) such that: ‚àë(Yacutual - Ypredicted)^2 is as small as possible.

square the errors because
->To avoid negative and positive errors canceling out
->To give more weight to larger errors
->To compute a smooth and differentiable function
```

```
Question 8: What is the purpose of the R-squared metric in regression analysis?
Answer:R-squared (R¬≤) is a metric used to measure how well a regression model explains the variability in the dependent variable.

Purpose of R-squared:

It tells us how much of the variation in the output (Y) is explained by the input variable(s) in the model.

It shows the goodness of fit of the regression line.

Interpretation:

R¬≤ ranges from 0 to 1:

0 ‚Üí Model explains none of the variation

1 ‚Üí Model explains all of the variation

Higher R¬≤ ‚Üí Better model performance

Example:
If R¬≤ = 0.80, it means 80% of the variation in Y is explained by X.
```


````python
#9.Question 9: Write Python code to fit a simple linear regression model using scikit-learn
# and print the slope and intercept.

import numpy as np
from sklearn.linear_model import LinearRegression

# Example data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 5, 4, 5])                 # Dependent variable

# Create and fit model
model = LinearRegression()
model.fit(X, y)

# Print coefficients
print("Slope (Œ≤1):", model.coef_[0])
print("Intercept (Œ≤0):", model.intercept_)

# Predict value
pred = model.predict([[6]])
print("Predicted value for X=6:", pred[0])


````
