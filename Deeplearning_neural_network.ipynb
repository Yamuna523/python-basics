Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?
A **Perceptron** is the simplest form of an **artificial neural network** and is a **binary linear classifier**.

--> It is a foundational neural model that performs linear classification, but its inability to model non-linear relationships led to the development of **multi-layer neural networks** and modern **deep learning architectures**.


**Components:**

1. **Input features**: (x_1, x_2, ..., x_n)
2. **Weights**: (w_1, w_2, ..., w_n)
3. **Bias**: (b)
4. **Weighted sum**:
   [
   z = \sum_{i=1}^{n} w_i x_i + b
   ]
   
5. **Activation function**
   [
   y =
   \begin{cases}
   1, & \text{if } z \ge 0 \
   0, & \text{otherwise}
   \end{cases}
   ]

---

### **Functioning of a Perceptron**

1. Inputs are multiplied by corresponding weights.
2. The weighted inputs are summed and a bias is added.
3. The result passes through an activation function.
4. The output is a class label (0 or 1).
5. **Training** adjusts weights using the **Perceptron Learning Rule**:
   [
   w_i = w_i + \eta (y_{true} - y_{pred}) x_i
   ]
   where ( \eta ) is the learning rate.

---

### **Limitations of a Perceptron**

1. **Only handles linearly separable data**

   * Cannot solve problems like **XOR**.

2. **Single-layer model**

   * No hidden layers → limited representational power.

3. **Binary output only**

   * Cannot handle multi-class problems directly.

4. **Hard threshold activation**

   * Step function is non-differentiable → limits optimization.

5. **Sensitive to noise**

   * Small changes in input can flip output.




Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

Purpose of Activation Functions

An activation function introduces non-linearity into a neural network. Without activation functions, a neural network would behave like a linear model, regardless of the number of layers, and would be unable to learn complex patterns.

Key purposes:

Enable learning of non-linear relationships

Control the output range of neurons

Help with gradient-based optimization

Decide whether a neuron should be activated or not

Sigmoid vs Tanh vs ReLU:-

->Sigmoid

Formula: 1 / (1 + e⁻ˣ)

Output Range: (0, 1)

Zero-centered: No

Vanishing Gradient: Yes

Computational Cost: High

Sparsity: No

Common Usage: Output layer for binary classification

->Tanh

Formula: tanh(x)

Output Range: (−1, 1)

Zero-centered: Yes

Vanishing Gradient: Yes (less severe than sigmoid)

Computational Cost: High

Sparsity: No

Common Usage: Hidden layers in older neural networks

->ReLU

Formula: max(0, x)

Output Range: [0, ∞)

Zero-centered: No

Vanishing Gradient: No (for positive inputs)

Computational Cost: Low

Sparsity: Yes

Common Usage: Hidden layers in modern deep neural networks



Question 4: What is the difference between Loss function and Cost function in neural
networks? Provide examples.
**Answer:**

**Loss Function**

* A loss function measures the error for a single training example.
* It tells how far the model’s prediction is from the true value for one data point.
* Used during training to compute gradients for each sample (or mini-batch).

**Examples:**

* **Mean Squared Error (single sample):**
  [
  L(y, \hat{y}) = (y - \hat{y})^2
  ]
* **Binary Cross-Entropy (single sample):**
  [
  L(y, \hat{y}) = -[y \log(\hat{y}) + (1-y)\log(1-\hat{y})]
  ]

---

**Cost Function**

* A **cost function** is the **aggregate (average or sum) of losses over the entire dataset**.
* It represents the **overall performance of the model**.
* Optimization algorithms (Gradient Descent, Adam) aim to **minimize the cost function**.

**Examples:**

* **Mean Squared Error (dataset):**
  [
  J = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2
  ]
* **Average Cross-Entropy Loss:**
  [
  J = \frac{1}{N} \sum_{i=1}^{N} L(y_i, \hat{y}_i)
  ]



Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.


**Answer:**

### **Role of Optimizers in Neural Networks**

An **optimizer** updates the **weights and biases** of a neural network to **minimize the loss/cost function**. It controls **how fast** and **how accurately** the model learns.

**Key roles:**

* Minimize training error
* Control learning rate and convergence speed
* Handle noisy or sparse gradients
* Improve training stability for deep networks

---

### **Comparison: Gradient Descent vs Adam vs RMSprop**

**1. Gradient Descent (GD)**

* Updates weights using the gradient of the loss function.
* Variants: **Batch GD, Stochastic GD (SGD), Mini-batch GD**

**Update rule:**
[
w = w - \eta \nabla J(w)
]

**Pros:**

* Simple and easy to understand
* Low memory usage

**Cons:**

* Slow convergence
* Sensitive to learning rate
* Gets stuck in local minima or saddle points

**Usage:** Small datasets, theoretical understanding

---

**2. RMSprop**

* Uses **adaptive learning rates** for each parameter.
* Maintains an **exponentially weighted moving average** of squared gradients.

**Update rule:**
[
v_t = \beta v_{t-1} + (1-\beta)(\nabla J)^2
]
[
w = w - \frac{\eta}{\sqrt{v_t + \epsilon}} \nabla J
]

**Pros:**

* Faster convergence than SGD
* Works well for **non-stationary** objectives
* Handles exploding/vanishing gradients better

**Cons:**

* No bias correction
* Slightly more memory usage

**Usage:** RNNs, deep networks

---

**3. Adam (Adaptive Moment Estimation)**

* Combines **Momentum + RMSprop**.
* Maintains moving averages of **gradients (1st moment)** and **squared gradients (2nd moment)**.

**Update rule:**
[
m_t = \beta_1 m_{t-1} + (1-\beta_1)\nabla J
]
[
v_t = \beta_2 v_{t-1} + (1-\beta_2)(\nabla J)^2
]

**Pros:**

* Fast convergence
* Bias correction included
* Works well with sparse gradients
* Default choice in practice

**Cons:**

* Can sometimes generalize worse than SGD
* More memory consumption

**Usage:** Most deep learning applications

````python
# Question 6: Write a Python program to implement a single-layer perceptron from
# scratch using NumPy to solve the logical AND gate.

import numpy as np

# AND gate dataset
X = np.array([
    [1, 1],
    [1, 0],
    [0, 1],
    [0, 0]
])

y = np.array([1, 0, 0, 0])
# Initialize weights and bias
weights = np.zeros(X.shape[1])
bias = 0
learning_rate = 0.1
epochs = 10

# Step activation function
def step_function(x):
    return 1 if x >= 0 else 0

# Training the perceptron
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = step_function(linear_output)

        # Update rule
        error = y[i] - y_pred
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Prediction: {y_pred}, Error: {error}")

print("\nFinal Weights:", weights)
print("Final Bias:", bias)

# Testing
print("\nTesting AND Gate:")
for i in range(len(X)):
    output = step_function(np.dot(X[i], weights) + bias)
    print(f"{X[i]} -> {output}")
````

````python
# Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions
# using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-10, 10, 100)

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plotting
plt.figure(figsize=(12, 4))

# Sigmoid
plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid)
plt.title("Sigmoid Activation")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.grid(True)

# ReLU
plt.subplot(1, 3, 2)
plt.plot(x, y_relu)
plt.title("ReLU Activation")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh)
plt.title("Tanh Activation")
plt.xlabel("x")
plt.ylabel("Tanh(x)")
plt.grid(True)

plt.tight_layout()
plt.show()
````

````python
# Question 8: Use Keras to build and train a simple multilayer neural network on the
# MNIST digits dataset. Print the training accuracy.

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset(multi neural network)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize data
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.1
)

# Print training accuracy
print("Training Accuracy:", history.history['accuracy'][-1])
````

````python
# Question 9: Visualize the loss and accuracy curves for a neural network model trained
# on the Fashion MNIST dataset. Interpret the training behavior.

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Build the model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)

# Plot Loss
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
plt.grid(True)
plt.show()

# Plot Accuracy
plt.figure()
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.legend()
plt.grid(True)
plt.show()

````

```
 Question 10: You are working on a project for a bank that wants to automatically detect
 fraudulent transactions. The dataset is large, imbalanced, and contains structured
 features like transaction amount, merchant ID, and customer location. The goal is to
 classify each transaction as fraudulent or legitimate.
 Explain your real-time data science workflow:
 ● How would you design a deep learning model (perceptron or multilayer NN)?
 ● Which activation function and loss function would you use, and why?
 ● How would you train and evaluate the model, considering class imbalance?
 ● Which optimizer would be suitable, and how would you prevent overfitting?

A.  **Fraud Detection – Real-Time Deep Learning Workflow**

### **1️ Model Design (Perceptron vs Multilayer Neural Network)**

* A **single-layer perceptron is insufficient** because fraud patterns are **non-linear and complex**.
* I would design a **Multilayer Neural Network (MLP)** for structured transaction data.

**Proposed Architecture:**

* **Input layer:** Transaction amount, merchant ID (embedded), location, time features
* **Hidden layers:** 2–4 dense layers (ReLU activation)
* **Output layer:** 1 neuron (Sigmoid activation)

**Why MLP?**

* Captures non-linear relationships
* Works well for tabular financial data
* Efficient for real-time inference

---

### **2️ Activation Function & Loss Function**

#### **Activation Functions**

* **Hidden layers:** `ReLU`

  * Fast computation
  * Avoids vanishing gradient
* **Output layer:** `Sigmoid`

  * Produces probability (0–1)
  * Ideal for binary classification (fraud / non-fraud)

#### **Loss Function**

* **Binary Cross-Entropy**
  [ L = -[y\log(p) + (1-y)\log(1-p)] ]

**Why?**

* Penalizes confident wrong predictions heavily
* Works well with probabilistic outputs
* Standard for fraud detection problems

---

### **3️ Training & Evaluation (Handling Class Imbalance)**

#### **Handling Imbalance**

* Fraud is rare → typically <1%
* Techniques:

  * **Class weights** (higher penalty for fraud class)
  * **SMOTE / undersampling** (during experimentation)
  * **Threshold tuning** (not default 0.5)

#### **Evaluation Metrics (NOT accuracy)**

* **Precision** → avoid false fraud alerts
* **Recall** → catch actual fraud
* **F1-score** → balance precision & recall
* **ROC-AUC / PR-AUC** → model discrimination

✔️ Priority depends on business:

* Bank prefers **high recall** (don’t miss fraud)
* Later optimize precision to reduce false alerts

---

### **4️. Optimizer & Overfitting Prevention**

#### **Optimizer**

* **Adam**

  * Adaptive learning rate
  * Fast convergence
  * Handles sparse gradients well
  * Ideal for large-scale financial data

#### **Overfitting Prevention**

* **Dropout layers** (e.g., 0.3–0.5)
* **L2 regularization**
* **Early stopping** on validation loss
* **Feature scaling & normalization**
* **Cross-validation on time-based splits**

---

### **5️. Real-Time Deployment Considerations**

* Low-latency inference (sub-100 ms)
* Model served via **REST / gRPC**
* Continuous retraining (concept drift)
* Monitor:

  * Fraud rate
  * False positives
  * Data drift

````python
#import libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, roc_auc_score

#sample transaction data
# Example structured features
X = pd.DataFrame({
    "amount": np.random.rand(10000) * 1000,
    "merchant_id": np.random.randint(1, 100, 10000),
    "location": np.random.randint(1, 50, 10000),
    "time_gap": np.random.rand(10000) * 10
})

# Highly imbalanced target (fraud = 1)
y = np.random.choice([0, 1], size=10000, p=[0.98, 0.02])



X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


#handle class imbalance

class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)

class_weight_dict = {
    0: class_weights[0],
    1: class_weights[1]
}

print(class_weight_dict)


# build multi neural network
model = Sequential([
    Dense(128, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.4),
    Dense(64, activation="relu"),
    Dropout(0.3),
    Dense(1, activation="sigmoid")
])


#compile model
model.compile(
    optimizer="adam",                     # Optimizer
    loss="binary_crossentropy",            # Loss function
    metrics=["accuracy"]
)

#train model with overfitting control

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=256,
    validation_split=0.2,
    class_weight=class_weight_dict,
    callbacks=[early_stop],
    verbose=1
)


y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_prob))

````
