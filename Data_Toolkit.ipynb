4.Explain the use of the groupby() method in Pandas?
A.In Pandas, the groupby() method is used to split data into groups based on
  the values in one or more columns, and then apply some operation (like aggregation, transformation, or filtering) on those groups

  Steps of groupby():
  ->Split – The data is split into groups based on a column (or multiple columns).
  ->Apply – A function (like sum(), mean(), count(), etc.) is applied to each group.
  ->Combine – The results are combined into a new DataFrame.
   ex:-
  import pandas as pd

  data = {
      'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
      'Employee': ['chari', 'srikanth', 'elizbeth', 'daniel', 'yamuna', 'krishna'],
      'Salary': [60000, 65000, 58000, 62000, 70000, 72000]
  }

  df = pd.DataFrame(data)
  print("Original DataFrame:\n", df)

  # Group by Department and calculate average salary
  grouped = df.groupby('Department')['Salary'].mean()
  print("\nAverage Salary by Department:\n", grouped)
  output:
    Original DataFrame:
        Department  Employee  Salary
      0         IT     chari   60000
      1         IT  srikanth   65000
      2         HR  elizbeth   58000
      3         HR    daniel   62000
      4    Finance    yamuna   70000
      5    Finance   krishna   72000

      Average Salary by Department:
      Department
      Finance    71000.0
      HR         60000.0
      IT         62500.0
      Name: Salary, dtype: float64

17. Explain the difference between apply() and map() in Pandas?
A. map:
   --> Works only on Pandas Series (one column).
   --> Applies a function, dictionary, or Series elementwise to transform values.
   --> Best for element-wise transformations.
   -->Performance wise slightly faster
   -->It is uses for value mapping
   example:
    import pandas as pd

    s = pd.Series([1, 2, 3, 4])

    # Using map with function
    print(s.map(lambda x: x**2))

    # Using map with dictionary
    print(s.map({1: 'A', 2: 'B', 3: 'C'}))

   apply:
    -> Works on both Series and DataFrame.
    -> It Can apply a function to:
        Each element (Series)
        Each row/column (DataFrame, depending on axis)
    -> More flexible and powerful than map().
    ->slightly slower compared to map
    ->using apply we can use complex functions

    example:-
    1) s = pd.Series([1, 2, 3, 4])
    print(s.apply(lambda x: x**2))

    2) df = pd.DataFrame({
        'A': [1, 2, 3],
        'B': [10, 20, 30]
    })

    # Apply function to each column
    print(df.apply(sum, axis=0))

 work with dates, times, and indexed temporal data compared to raw Python. It
  provides high-level tools for parsing, indexing, resampling, and performing operations on time-based data.
  1.Datetime Handling

     dates = pd.to_datetime(["2025-01-01", "2025-01-02", "2025-01-03"])

  2.DatetimeIndex & Time-based Indexing

      dates = pd.date_range("2025-01-01", periods=3, freq="D")

  3.Resampling & Frequency Conversion

    example:
    rng = pd.date_range('2025-01-01', periods=6, freq='T')  # 6 minutes frequency
    ts = pd.Series(range(len(rng)), index=rng)
    print(ts)

  4.Shifting & Lagging Data
   example :
    print(ts.shift(2))
  5.Rolling & Expanding Windows
  #example
  print(ts.rolling(window=3).mean())
  print(ts.expanding(min_periods=1).mean())
  6.Time Zone Handling
  print(dates.tz_localize('UTC'))
